Plotting Cross Validation Metrics

```python
#General Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Metrics from RF Models

```python
cross_df = pd.read_csv('cross_val_results.csv')
```

Training

```python
fig_dims = (20, 10)

box_training, ax = plt.subplots(figsize=fig_dims)
sns.boxplot(data=cross_df[['cv 02 training', 'cv 03 training', 'cv 04 training', 'cv 05 training',
                           'cv 06 training', 'cv 07 training', 'cv 08 training',
                           'cv 09 training', 'cv 10 training']])
#ax.set(xlabel= "fold", ylabel= "R2 score")
ax.set_xlabel("fold", fontsize = 12)
ax.set_ylabel('r2 score', fontsize = 12)
plt.savefig('Cross Val Boxplot - Training', dpi = 600)
```

Testing

```python
fig_dims = (20, 10)

box_testing, ax = plt.subplots(figsize=fig_dims)
sns.boxplot(data=cross_df[['cv 02 testing', 'cv 03 testing', 'cv 04 testing', 'cv 05 testing',
                           'cv 06 testing', 'cv 07 testing', 'cv 08 testing',
                           'cv 09 testing', 'cv 10 testing']])
#ax.set(xlabel= "fold", ylabel= "R2 score")
ax.set_xlabel("fold", fontsize = 12)
ax.set_ylabel('r2 score', fontsize = 12)
plt.savefig('Cross Val Boxplot - testing', dpi = 600)
```
